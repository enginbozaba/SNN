Figure 1. The conflict between spike-timing-dependent plasticity (STDP) and spike synchronization in Hebbian cell assemblies. (A) In standard spike pair-based (doublet) STDP models, the weight change w is a function F(t) of the time lag t: = tpost  tpre between presynaptic and postsynaptic spike. For positive time lags t > 0 (pre-before-post) STDP models predict long-term potentiation (LTP) with w > 0. For negative time lags t < 0 (post-before-pre) STDP models predict long-term depression (LTD) with w < 0.
